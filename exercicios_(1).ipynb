Remover duplicados da lista.

```python
print('Remover duplicados dessa lista 11,11,12,15,20,20,21,23')

def remove_duplicados(a):
    lista_ret = []
    for i in a:
        if i not in lista_ret:
            lista_ret.append(i)
    lista_ret.sort()
    return lista_ret

teste = [11,11,12,15,20,20,21,23]

resultado = remove_duplicados(teste)
print(f'Vai ficar nessa sequencia {resultado}')
```

## Exercício 34
Gerar lista de primos até N.

```python
print('Gerar lista de primos ate N.')

numero = int(input('Digite o numero inteiro positivo para verificar se é primo '))

def lista_de_primos(a):
    primos = []
    for num in range(2, a + 1):
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                break
        else:
            primos.append(num)
    return primos
print(f'a lista referente ao valor digitado {numero} composto somente de numeros primos é {lista_de_primos(numero)}')


```

## Exercício 59
Crie uma função que remove duplicados de uma lista.

```python
print('Crie uma função que remove duplicados da lista 11,11,12,15,20,20,21,23')

def remove_duplicados(a):
    lista_retorno = []
    for i in a:
        if i not in lista_retorno:
            lista_retorno.append(i)
    lista_retorno.sort()
    return lista_retorno

teste = [11,11,12,15,20,20,21,23]

resultado = remove_duplicados(teste)
print(f'Vai ficar nessa sequencia {resultado}')
```

```python
print('Crie uma função que retorna apenas números positivos')
print('Lista com numeros de exemplo -11,11,-12,15,-20,20,21,-23')

def somente_positivos(a):
    lista_retorno = []
    for i in a:
        if i > 0:
            lista_retorno.append(i)
    lista_retorno.sort()
    return lista_retorno

teste = [-11,11,-12,15,-20,20,21,-23]

resultado = somente_positivos(teste)
print(f'Vai ficar nessa sequencia {resultado}')
```
